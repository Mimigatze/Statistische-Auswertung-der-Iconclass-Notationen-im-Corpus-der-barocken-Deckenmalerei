```python
# Iconclass符号在巴洛克天顶画语料库(CbDD)中的统计分析
# 作者: [您的姓名]
# 日期: [当前日期]

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import re

# 设置matplotlib字体和图表样式
import matplotlib
import matplotlib.pyplot as plt

# 不使用中文，使用英文标签，避免字体问题
plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams["axes.labelsize"] = 10  # 较小的标签文字
plt.rcParams["axes.titlesize"] = 12  # 较小的标题文字
plt.rcParams["legend.fontsize"] = 9  # 较小的图例文字

# 设置matplotlib风格为更现代的样式
plt.style.use('seaborn-v0_8-whitegrid')

# 检查系统字体
def check_font_support():
    """检查当前系统上可用的字体"""
    print("正在检查系统字体...")
    
    # 列出可用字体
    from matplotlib.font_manager import fontManager
    available_fonts = sorted([f.name for f in fontManager.ttflist])
    print(f"系统上可用的字体数量: {len(available_fonts)}")
    print(f"部分可用字体: {available_fonts[:10]}...")
    
    # 建议使用的字体
    recommended_fonts = ['Arial', 'DejaVu Sans', 'Verdana', 'Times New Roman', 'Calibri']
    available_recommended = [f for f in recommended_fonts if f in available_fonts]
    print(f"推荐使用的可用字体: {available_recommended}")
    
    # 使用第一个可用的推荐字体
    if available_recommended:
        selected_font = available_recommended[0]
        matplotlib.rcParams['font.family'] = 'sans-serif'
        matplotlib.rcParams['font.sans-serif'] = [selected_font]
        print(f"已选择字体: {selected_font}")
    else:
        print("未找到推荐字体，使用系统默认字体")

# 载入数据
def load_data():
    # 载入房间和Iconclass数据
    rooms_df = pd.read_csv('rooms_iconclasses.csv')
    # 载入Iconclass名称数据
    iconclass_names_df = pd.read_csv('iconclass_names.csv')



    
    # 创建映射字典
    iconclass_names_dict = dict(zip(iconclass_names_df['iconclass_id'], iconclass_names_df['description']))
    
    return rooms_df, iconclass_names_dict

# 数据清理函数
def clean_data(df):
    # 检查缺失值
    print("缺失值统计:")
    print(df.isnull().sum())
    
    # 确保iconclassses_level_0和iconclassses_level_1列的类型是字符串
    for col in ['iconclassses_level_0', 'iconclassses_level_1']:
        # 将数值型转换为字符串
        mask = df[col].notna() & df[col].apply(lambda x: isinstance(x, (int, float)))
        if mask.any():
            df.loc[mask, col] = df.loc[mask, col].astype(str)
    
    return df

# 任务1: 分析Iconclass符号在Level 0和Level 1层级上的频率
def analyze_iconclass_levels(df, iconclass_names):
    # 处理Level 0数据
    level0_data = process_multi_value_column(df, 'iconclassses_level_0')
    
    # 处理Level 1数据
    level1_data = process_multi_value_column(df, 'iconclassses_level_1')
    
    # 添加名称信息
    level0_with_names = add_names_to_frequency(level0_data, iconclass_names)
    level1_with_names = add_names_to_frequency(level1_data, iconclass_names)
    
    return level0_with_names, level1_with_names

# 处理多值列(逗号分隔的值)
def process_multi_value_column(df, column_name):
    # 过滤掉null值和空字符串
    valid_data = df[df[column_name].notna() & (df[column_name] != '')]
    
    # 拆分逗号分隔的值并展平为一个列表
    all_values = []
    for value in valid_data[column_name]:
        try:
            values = str(value).split(',')
            all_values.extend(values)
        except:
            print(f"处理值时出错: {value}, 类型: {type(value)}")
    
    # 统计每个值的频率
    value_counts = Counter(all_values)
    
    # 转换为DataFrame并排序
    freq_df = pd.DataFrame({
        'value': list(value_counts.keys()),
        'count': list(value_counts.values())
    })
    freq_df = freq_df.sort_values('count', ascending=False).reset_index(drop=True)
    
    # 计算百分比
    total = freq_df['count'].sum()
    freq_df['percentage'] = (freq_df['count'] / total * 100).round(2)
    
    return freq_df

# 添加Iconclass名称到频率数据
def add_names_to_frequency(freq_df, iconclass_names):
    # 添加名称列
    freq_df['name'] = freq_df['value'].map(lambda x: iconclass_names.get(x, 'Unbekannt'))
    
    # 重排列顺序
    freq_df = freq_df[['value', 'name', 'count', 'percentage']]
    
    return freq_df

# 可视化Level 0频率分布
def visualize_level0(level0_df, title_prefix=""):
    # 准备数据
    level0_top = level0_df.head(10).copy()
    
    # 条形图 - 频率
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # 绘制条形图 - 直接使用matplotlib而不是seaborn
    bars = ax.bar(range(len(level0_top)), level0_top['count'])
    
    # 设置标题和标签
    ax.set_title(f'{title_prefix}Iconclass Level 0 Frequency Distribution')
    ax.set_xlabel('Iconclass Level 0 Category')
    ax.set_ylabel('Count')
    
    # 明确设置刻度位置
    ax.set_xticks(range(len(level0_top)))
    
    # 使用简化的标签
    simple_labels = [f"{row.value}" for row in level0_top.itertuples()]
    ax.set_xticklabels(simple_labels)
    
    # 添加数值标签
    for i, bar in enumerate(bars):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 50,
                f"{int(height)}\n({level0_top.iloc[i]['percentage']}%)",
                ha='center', va='bottom')
    
    # 添加图例，显示类别说明
    legend_text = "\n".join([f"{row.value}: {row.name}" for row in level0_top.itertuples()])
    plt.figtext(0.01, 0.01, legend_text, wrap=True, fontsize=9, 
                bbox=dict(facecolor='white', alpha=0.8))
    
    plt.tight_layout()
    plt.savefig('level0_frequency_bar.png')
    plt.show()
    
    # 饼图 - 百分比 (改进标签位置，防止重叠)
    fig, ax = plt.subplots(figsize=(12, 12))
    
    # 控制标签的位置
    wedges, texts, autotexts = ax.pie(
        level0_df['percentage'], 
        autopct='%1.1f%%', 
        startangle=90,
        pctdistance=0.85,  # 将百分比标签移到内部
        wedgeprops=dict(width=0.5)  # 创建圆环状饼图
    )
    
    # 设置标题
    ax.set_title(f'{title_prefix}Iconclass Level 0 Percentage Distribution')
    
    # 创建单独的图例，而不是在饼图上直接标注
    legend_labels = [f"{row.value}: {row.name} ({row.percentage}%)" for row in level0_df.itertuples()]
    ax.legend(wedges, legend_labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    plt.tight_layout()
    plt.savefig('level0_percentage_pie.png')
    plt.show()
    
    # 表格输出
    print(f"{title_prefix}Iconclass Level 0 Statistics:")
    print(level0_df.to_string(index=False))

# 可视化Level 1频率分布
def visualize_level1(level1_df, title_prefix=""):
    # 水平条形图 - 频率 (只展示前15个)
    level1_top15 = level1_df.head(15).copy()
    
    # 创建图形和轴
    fig, ax = plt.subplots(figsize=(14, 10))
    
    # 绘制水平条形图
    bars = ax.barh(range(len(level1_top15)), level1_top15['count'])
    
    # 设置标题和标签
    ax.set_title(f'{title_prefix}Iconclass Level 1 Frequency Distribution (Top 15)')
    ax.set_xlabel('Count')
    ax.set_ylabel('Iconclass Level 1 Category')
    
    # 设置y轴刻度位置
    ax.set_yticks(range(len(level1_top15)))
    
    # 使用简化的标签
    simple_labels = [f"{row.value}" for row in level1_top15.itertuples()]
    ax.set_yticklabels(simple_labels)
    
    # 添加数值标签
    for i, bar in enumerate(bars):
        width = bar.get_width()
        ax.text(width + 10, bar.get_y() + bar.get_height()/2,
                f"{int(width)} ({level1_top15.iloc[i]['percentage']}%)",
                ha='left', va='center')
    
    # 添加图例，显示类别说明
    legend_text = "\n".join([f"{row.value}: {row.name}" for row in level1_top15.itertuples()])
    plt.figtext(0.01, 0.01, legend_text, wrap=True, fontsize=9,
                bbox=dict(facecolor='white', alpha=0.8))
    
    plt.tight_layout()
    plt.savefig('level1_frequency_bar.png')
    plt.show()
    
    # 饼图 - 百分比 (只展示前8个，其余归为"其他")
    top8 = level1_df.head(8).copy()
    others_sum = level1_df.iloc[8:]['percentage'].sum()
    others_count = level1_df.iloc[8:]['count'].sum()
    
    # 创建饼图数据
    pie_data = pd.DataFrame({
        'value': list(top8['value']) + ['Others'],
        'name': list(top8['name']) + ['Other Categories'],
        'count': list(top8['count']) + [others_count],
        'percentage': list(top8['percentage']) + [others_sum]
    })
    
    # 创建图形
    fig, ax = plt.subplots(figsize=(12, 12))
    
    # 控制标签的位置
    wedges, texts, autotexts = ax.pie(
        pie_data['percentage'], 
        autopct='%1.1f%%', 
        startangle=90,
        pctdistance=0.85,  # 将百分比标签移到内部
        wedgeprops=dict(width=0.5)  # 创建圆环状饼图
    )
    
    # 设置标题
    ax.set_title(f'{title_prefix}Iconclass Level 1 Percentage Distribution (Top 8 + Others)')
    
    # 创建单独的图例，而不是在饼图上直接标注
    legend_labels = [f"{row.value}: {row.name} ({row.percentage:.1f}%)" for row in pie_data.itertuples()]
    ax.legend(wedges, legend_labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    plt.tight_layout()
    plt.savefig('level1_percentage_pie.png')
    plt.show()
    
    # 表格输出
    print(f"{title_prefix}Iconclass Level 1 Statistics (Top 20):")
    print(level1_df.head(20).to_string(index=False))
    
    # 饼图 - 百分比 (只展示前8个，其余归为"其他")
    top8 = level1_df.head(8).copy()
    others = pd.DataFrame([{
        'value': 'Andere',
        'name': 'Andere Kategorien',
        'count': level1_df.iloc[8:]['count'].sum(),
        'percentage': level1_df.iloc[8:]['percentage'].sum()
    }])
    pie_data = pd.concat([top8, others], ignore_index=True)
    
    plt.figure(figsize=(12, 12))
    plt.pie(pie_data['percentage'], 
            labels=[f"{row.value} - {row.name}\n({row.percentage:.1f}%)" for row in pie_data.itertuples()],
            autopct='%1.1f%%', startangle=90)
    plt.title(f'{title_prefix}Iconclass Level 1 Prozentuale Verteilung (Top 8 + Andere)')
    plt.axis('equal')
    plt.tight_layout()
    plt.savefig('level1_percentage_pie.png')
    plt.show()
    
    # 表格输出
    print(f"{title_prefix}Iconclass Level 1 统计结果 (前20类):")
    print(level1_df.head(20).to_string(index=False))

# 任务2: 不同时间段的Iconclass层级分布比较
def compare_time_periods(df, iconclass_names):
    # 创建两个时间段的数据子集
    period1 = df[(df['dating_start'] >= 1550) & (df['dating_end'] <= 1800)]
    period2 = df[(df['dating_start'] >= 1600) & (df['dating_end'] <= 1800)]
    
    print(f"1550-1800年时间段的记录数: {len(period1)}")
    print(f"1600-1800年时间段的记录数: {len(period2)}")
    
    # 分析两个时间段的Level 0分布
    level0_period1 = process_multi_value_column(period1, 'iconclassses_level_0')
    level0_period1 = add_names_to_frequency(level0_period1, iconclass_names)
    
    level0_period2 = process_multi_value_column(period2, 'iconclassses_level_0')
    level0_period2 = add_names_to_frequency(level0_period2, iconclass_names)
    
    # 分析两个时间段的Level 1分布
    level1_period1 = process_multi_value_column(period1, 'iconclassses_level_1')
    level1_period1 = add_names_to_frequency(level1_period1, iconclass_names)
    
    level1_period2 = process_multi_value_column(period2, 'iconclassses_level_1')
    level1_period2 = add_names_to_frequency(level1_period2, iconclass_names)
    
    return {
        'level0_1550_1800': level0_period1,
        'level1_1550_1800': level1_period1,
        'level0_1600_1800': level0_period2,
        'level1_1600_1800': level1_period2
    }

# 比较可视化
def visualize_time_comparison(period_data):
    # Level 0比较可视化
    visualize_level0(period_data['level0_1550_1800'], "1550-1800: ")
    visualize_level0(period_data['level0_1600_1800'], "1600-1800: ")
    
    # Level 1比较可视化
    visualize_level1(period_data['level1_1550_1800'], "1550-1800: ")
    visualize_level1(period_data['level1_1600_1800'], "1600-1800: ")
    
    # 创建比较图 - Level 0
    compare_top_categories(
        period_data['level0_1550_1800'], 
        period_data['level0_1600_1800'],
        'Iconclass Level 0',
        '1550-1800', 
        '1600-1800',
        5  # 比较前5个类别
    )
    
    # 创建比较图 - Level 1
    compare_top_categories(
        period_data['level1_1550_1800'], 
        period_data['level1_1600_1800'],
        'Iconclass Level 1',
        '1550-1800', 
        '1600-1800',
        8  # 比较前8个类别
    )

# 比较两个时期的顶级类别
def compare_top_categories(df1, df2, level_name, period1_name, period2_name, top_n=5):
    # 获取两个数据集中全部的前N个类别
    top_categories = pd.concat([df1.head(top_n)['value'], df2.head(top_n)['value']]).unique()
    
    # 创建比较数据框
    compare_data = []
    for category in top_categories:
        p1_row = df1[df1['value'] == category]
        p2_row = df2[df2['value'] == category]
        
        p1_percentage = p1_row['percentage'].values[0] if not p1_row.empty else 0
        p2_percentage = p2_row['percentage'].values[0] if not p2_row.empty else 0
        
        name = p1_row['name'].values[0] if not p1_row.empty else p2_row['name'].values[0]
        
        compare_data.append({
            'category': category,
            'name': name,
            period1_name: p1_percentage,
            period2_name: p2_percentage,
            'difference': p2_percentage - p1_percentage
        })
    
    compare_df = pd.DataFrame(compare_data)
    compare_df = compare_df.sort_values(by=period1_name, ascending=False)
    
    # 可视化比较 - 重写以避免警告
    fig, ax = plt.subplots(figsize=(12, 8))
    x = np.arange(len(compare_df))
    width = 0.35
    
    # 绘制条形图
    bar1 = ax.bar(x - width/2, compare_df[period1_name], width, label=period1_name)
    bar2 = ax.bar(x + width/2, compare_df[period2_name], width, label=period2_name)
    
    # 设置标题
    ax.set_title(f'{level_name} Vergleich zwischen {period1_name} und {period2_name}')
    
    # 明确设置x轴刻度位置和标签
    ax.set_xticks(x)
    
    # 为了处理可能的特殊字符，使用更简洁的标签
    simplified_labels = [f"{row.category}" for row in compare_df.itertuples()]
    ax.set_xticklabels(simplified_labels)
    
    # 在图表底部添加一个图例，解释每个类别代表什么
    category_legend = "\n".join([f"{row.category}: {row.name}" for row in compare_df.itertuples()])
    plt.figtext(0.1, 0.01, category_legend, wrap=True, fontsize=8)
    
    plt.xticks(rotation=45, ha='right')
    ax.legend()
    
    # 添加标签
    for i, v in enumerate(compare_df[period1_name]):
        ax.text(i - width/2, v + 0.5, f"{v:.1f}%", ha='center')
    
    for i, v in enumerate(compare_df[period2_name]):
        ax.text(i + width/2, v + 0.5, f"{v:.1f}%", ha='center')
    
    plt.tight_layout()
    plt.subplots_adjust(bottom=0.2)  # 为底部图例腾出空间
    plt.savefig(f'compare_{level_name.lower().replace(" ", "_")}_{period1_name}_{period2_name}.png')
    plt.show()
    
    # 打印完整表格以供参考
    print("\n比较表格:")
    print(compare_df.to_string(index=False))
    
    # 显示比较表格
    print(f"{level_name}类别在{period1_name}和{period2_name}之间的比较:")
    print(compare_df.to_string(index=False))

def main():
    # 检查字体支持
    check_font_support()
    
    # 加载数据
    print("\n正在加载数据...")
    rooms_df, iconclass_names = load_data()
    
    # 数据清洗
    print("\n正在清洗数据...")
    rooms_df = clean_data(rooms_df)
    
    # 任务1: 分析Iconclass符号在三个层级上的频率
    print("\n任务1: 分析Iconclass符号在Level 0和Level 1层级上的频率")
    level0_freq, level1_freq = analyze_iconclass_levels(rooms_df, iconclass_names)
    
    # 可视化任务1结果
    try:
        visualize_level0(level0_freq)
        visualize_level1(level1_freq)
        print("任务1可视化完成")
    except Exception as e:
        print(f"任务1可视化时出错: {str(e)}")
    
    # 任务2: 不同时间段的Iconclass层级分布比较
    print("\n任务2: 不同时间段的Iconclass层级分布比较")
    period_data = compare_time_periods(rooms_df, iconclass_names)
    try:
        visualize_time_comparison(period_data)
        print("任务2可视化完成")
    except Exception as e:
        print(f"任务2可视化时出错: {str(e)}")

if __name__ == "__main__":
    main()
```
